**1. Chicago North shore - spring(5)**

```python
url = 'https://www.visitchicagonorthshore.com/Blog/celebrate_spring'
r = requests.get(url)
```

```python
print(r.status_code)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')

print(soup.title)
```

```python
script = soup.find_all('div')[27]
print(script.prettify())
```

```python
sections = script.find_all('p', recursive=False)
# print(sections)
```

```python
titles = []
descriptions = []
images = []
```

```python
for section in sections:
    title = section.find('a', target='_blank')
    description = section.find_all('br')[1].next_sibling.strip() if len(section.find_all('br')) > 1 else ''
    image = section.find_previous('img')['src'] if section.find_previous('img') else ''
    
    if title:
        titles.append(title.text.strip())
        descriptions.append(description)
        images.append(image)
```

```python
df = pd.DataFrame({
    'title': titles,
    'description': descriptions,
    'image': images
})
```

```python
df = df.drop(df.index[0])
df['title'] = df['title'].str.replace('Enjoy: ', '', regex=False)
df['Source'] = 'Chicago North Shore'
df['category'] = 'spring'
df.head(6)
```

```python
df_1 = df
df_1.shape[0]
```

**6. Enjoy Illinois - Scienic Byways (7) | 134**

```python
url = 'https://www.enjoyillinois.com/things-to-do/illinois-scenic-byways/'
r = requests.get(url)
```

```python
print(r.status_code)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')

print(soup.title)
```

```python
script = soup.find_all('main')[0]
print(script.prettify())
```

```python
sections = script.find_all('section')
print(sections)
```

```python
titles = []
descriptions = []
images = []

base_url = "https://www.enjoyillinois.com"

for section in sections:
    # Extract the title
    title_tag = section.find('h2')
    title = title_tag.get_text(strip=True) if title_tag else 'No title found'
    
    # Extract the image URL
    figure_tag = section.find('figure')
    if figure_tag:
        img_tag = figure_tag.find('img')
        if img_tag:
            image_url = img_tag.get('data-src', img_tag.get('src', ''))
            if image_url.startswith('/'):
                image_url = base_url + image_url
        else:
            image_url = 'No image found'
    else:
        image_url = 'No image found'
    
    # Extract the description
    content_box_tag = section.find('div', class_='content-box')
    if content_box_tag:
        description_tag = content_box_tag.find('p')
        description = description_tag.get_text(strip=True) if description_tag else 'No description found'
    else:
        description = 'No description found'
    
    # Append to the lists
    titles.append(title)
    descriptions.append(description)
    images.append(image_url)
```

```python
df = pd.DataFrame({
    'title': titles,
    'description': descriptions,
    'image': images,
    'Source': 'Enjoy Illinois',
    'category': 'Scienic Byways'
})
df = df[df['title'] != 'No title found']
df.head()
```

```python
df_6 = df
df_6.shape[0]
```

```python
import regex as re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
punctuation = '!”$%&\’()*+,-./:;<=>?[\\]^_`{|}~•@©—'

def remove_links(text):
    """Takes a string and removes web links from it"""
    text = re.sub(r'http\S+', '', text)  # remove http links
    text = re.sub(r'bit.ly/\S+', '', text)  # remove bitly links
    text = text.strip('[link]')  # remove [links]
    text = text.strip()
    text = re.sub(r'pic.twitter\S+', '', text)
    return text


def remove_tags(text):
    remove = re.compile(r'')
    return re.sub(remove, '', text)


def remove_users(text):
    """Takes a string and removes retweet and @user information"""
    text = re.sub('(RT\s@[A-Za-z]+[A-Za-z0-9-_]+)', '', text)  # remove re-tweet
    text = re.sub('(@[A-Za-z]+[A-Za-z0-9-_]+)', '', text)  # remove tweeted at
    return text


def remove_hashtags(text):
    """Takes a string and removes any hash tags"""
    text = re.sub('(#[A-Za-z]+[A-Za-z0-9-_]+)', '', text)  # remove hash tags
    return text


def remove_av(text):
    """Takes a string and removes AUDIO/VIDEO tags or labels"""
    text = re.sub('VIDEO:', '', text)  # remove 'VIDEO:' from start of tweet
    text = re.sub('AUDIO:', '', text)  # remove 'AUDIO:' from start of tweet
    return text


def basic_clean(text):
    """Main master function to clean tweets only without tokenization or removal of stopwords"""
    text = remove_users(text)
    text = remove_links(text)
    text = remove_hashtags(text)
    text = remove_tags(text)
    text = remove_av(text)
    text = text.lower()  # lower case
    text = re.sub('[' + punctuation + ']+', ' ', text)  # strip punctuation
    text = re.sub('\s+', ' ', text)  # remove double spacing
    text = re.sub('([0-9]+)', '', text)  # remove numbers
    text = re.sub('📝 …', '', text)
    text = re.sub('–', '', text)
    return text


def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    return [x for x in words if x not in stop_words]


def lemmatize_word(text):
    wordnet = WordNetLemmatizer()
    return " ".join([wordnet.lemmatize(word) for word in text])
```

```python
df['text_cleaned'] = df['description'].apply(remove_stopwords)
df['text_cleaned'] = df['text_cleaned'].apply(lemmatize_word)
# df['text_cleaned_token'] = df['text_cleaned_token'].apply(remove_stopwords)

df.to_csv('second_scrap_ak.csv', index=False)
```
